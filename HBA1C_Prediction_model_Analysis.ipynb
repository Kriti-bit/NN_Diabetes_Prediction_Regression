```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
##### Creating Tensors
X_train=torch.tensor(X_train)
X_test=torch.tensor(X_test)
```

```python
y_train=torch.FloatTensor(y_train)
y_test=torch.FloatTensor(y_test)
```

```python
class ANN_Model(nn.Module):
    def __init__(self,input_features=4,hidden1=25,hidden2=25,out_features=1):
        super().__init__()
        self.f_connected1=nn.Linear(input_features,hidden1)
        self.f_connected2=nn.Linear(hidden1,hidden2)
        self.out=nn.Linear(hidden2,out_features)
    def forward(self,x):
        x=F.relu(self.f_connected1(x))
        x=F.relu(self.f_connected2(x))
        x=self.out(x)
        return x
```

```python
torch.manual_seed(20)
model = ANN_Model()
```

```python
### Backward Propagation:- Define the loss function and the optimizer
loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 0.01) # lr refers to the learning rate
```

```python
#   # Run the training loop
#   for epoch in range(0, 10): # 10 epochs at maximum
    
#     # Print epoch
#     print(f'Starting epoch {epoch+1}')
    
#     # Set current loss value
#     current_loss = 0.0
    
#     # Iterate over the DataLoader for training data
#     for i, data in enumerate(X_test):
      
#       # Get and prepare inputs
#       inputs, targets = data
#       inputs, targets = inputs.float(), targets.float()
#       targets = targets.reshape((targets.shape[0], 1))
      
#       # Zero the gradients
#       optimizer.zero_grad()
      
#       # Perform forward pass
#       outputs = mlp(inputs)
      
#       # Compute loss
#       loss = loss_function(outputs, targets)
      
#       # Perform backward pass
#       loss.backward()
      
#       # Perform optimization
#       optimizer.step()
      
#       # Print statistics
#       current_loss += loss.item()
#       if i % 10 == 0:
#           print('Loss after mini-batch %5d: %.3f' %
#                 (i + 1, current_loss / 500))
#           current_loss = 0.0

#   # Process is complete.
#   print('Training process has finished.')
    
    
epochs = 500
final_losses = []
for i in range(epochs):
    i = i + 1
    y_pred = model.forward(X_train.float())
    loss=loss_function(y_pred, y_train)
    final_losses.append(loss.item())
    if i%10==1:
        print("Epoch number: {} and the loss: {}".format(i, loss.item()))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
